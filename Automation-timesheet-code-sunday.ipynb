```python
# dataframe for sunday as weekend and Friday and Saturday as weekoff
time_sunday_weekend = pd.DataFrame()
```

```python
# defining the function that creates the dates of a month according to the month and the year name provided

def create_range(month, year):
    d = pd.to_datetime(f'{month} {year}', format='%B %Y')
    return pd.date_range(start = d, periods = d.daysinmonth)    
```

```python
#adjust the parameters in this step to get the desired output(for Sunday as weekend and Friday and Saturday as weekoff)
date_creator = create_range(month = 'November', year = 2023)

time_sunday_weekend["Date"] = date_creator

time_sunday_weekend["Resource"] = "Rajib"

time_sunday_weekend["Resource_Project_code"] = "Project_code"

time_sunday_weekend["P_A"] = 0
time_sunday_weekend["P_B"] = 0
time_sunday_weekend["P_C"] = 0
time_sunday_weekend["P_D"] = 0
time_sunday_weekend["P_E"] = 0
time_sunday_weekend["P_F"] = 0
time_sunday_weekend["P_G"] = 0
```

```python
time_sunday_weekend.head()
```

```python
# function to convert the date to string in dd-mm-yyyy format

def convert_date_standard_string(obj):
    return "{:%d-%m-%Y}".format(obj)
```

```python
#function to make customised weekend (for Friday and saturday as week-off)
def is_weekoff_weekend_sunday(date):
    if date.weekday() == 4 or date.weekday() == 5:
        return "Week-Off"
    elif date.weekday() == 6:
        return "Week-End"
    else:
        return "Working Day"

time_sunday_weekend['day-description'] = time_sunday_weekend['Date'].apply(is_weekoff_weekend_sunday)
```

```python
# for sunday weekend

time_sunday_weekend["Weekday Hours"] = 0
time_sunday_weekend["Weekend Hours"] = 0
time_sunday_weekend["Total Hours"]=0
time_sunday_weekend["Comments"]=" "
time_sunday_weekend["Weekend comments"]=" "

time_sunday_weekend['Date'] = time_sunday_weekend['Date'].apply(convert_date_standard_string)
```

```python
writer = pd.ExcelWriter('sunday_weekend.xlsx', engine = 'xlsxwriter', date_format='%d-%m-%Y')

time_sunday_weekend.to_excel(writer, sheet_name='hours', startrow=13,startcol=0, index=False)

workbook = writer.book
worksheet = writer.sheets['hours']
green = workbook.add_format({'color':'green'})
red = workbook.add_format({'color':'red'})




string = ['Weekoffs, Leaves, Holidays are marked in',red,' red.', ' Weekends are marked in', green, ' green.']
worksheet.write(0, 0, 'Note: ')
worksheet.write_rich_string(1, 0, 
                            *string)

# setting the border format
centre_border = workbook.add_format({'border': 1, "align":"middle"})
centre_border.set_text_wrap()

# # merging the columns to set the header
# centre_format = workbook.add_format(
#     {
#         "align":"middle"
#     }
# )

worksheet.write(3, 0, 'Project name and Tasks-')
worksheet.write(4, 0, 'P_A - This is the task for project A. [US Project code - US_A]')
worksheet.write(5, 0, 'P_B - This is the task for project B. [US Project code - US_B]')
worksheet.write(6, 0, 'P_C - This is the task for project C. [US Project code - US_C]')
worksheet.write(7, 0, 'P_D - This is the task for project D. [US Project code - US_D]')
worksheet.write(8, 0, 'P_E - This is the task for project E. [US Project code - US_E]')
worksheet.write(9, 0, 'P_F - This is the task for project F. [US Project code - US_F]')
worksheet.write(10, 0, 'P_G - This is the task for project G. [US Project code - US_G]')





# # creating the space for calculating the total number of hours for weekend and weekdays

worksheet.write(48, 1, 'Project', centre_border)
worksheet.write(48, 2, 'US-Project code',centre_border)
worksheet.write(48, 3, 'Total Weekday hours',centre_border)
worksheet.write(48, 4, 'Total Weekend hours', centre_border)
worksheet.write(48, 5, 'Total hours',centre_border)


# writing the project names
worksheet.write(49, 1, 'P_A')
worksheet.write(50, 1, 'P_B')
worksheet.write(51, 1, 'P_C')
worksheet.write(52, 1, 'P_D')
worksheet.write(53, 1, 'P_E')
worksheet.write(54, 1, 'P_F')
worksheet.write(55, 1, 'P_G')

# writing the US-Project codes

worksheet.write(49, 2, 'US_A')
worksheet.write(50, 2, 'US_B')
worksheet.write(51, 2, 'US_C')
worksheet.write(52, 2, 'US_D')
worksheet.write(53, 2, 'US_E')
worksheet.write(54, 2, 'US_F')
worksheet.write(55, 2, 'US_G')









# formatting the cells based off their values
worksheet.conditional_format('A15:P46', {'type':'formula', 'criteria':'=$K15="Week-Off"','format':red})
worksheet.conditional_format('A15:P46', {'type':'formula', 'criteria':'=$K15="Week-End"','format':green})
worksheet.conditional_format('A15:P46', {'type':'formula', 'criteria':'=$K15="Leave"','format':red})
worksheet.conditional_format('A15:P46', {'type':'formula', 'criteria':'=$K15="Holiday"','format':red})


# aligning the values in the middle

centre_format = workbook.add_format()
centre_format.set_align('vcentre')

worksheet.set_column('A:XFD', None, centre_format)

# setting the column width

worksheet.set_column(0,0, 12.5)
worksheet.set_column(1,13, 20)
worksheet.set_column(14, 15, 33.5)



# displaying the total hours in weekend and weekdays and also including data for Half-day leaves

# for project A
worksheet.write_formula('D50','=SUMIF(K15:K45,"Working Day", D15:D45)' )
worksheet.write_formula('E50','=SUMIF(K15:K45,"Week-End", D15:D45)' )

# for project B
worksheet.write_formula('D51','=SUMIF(K15:K45,"Working Day", E15:E45)' )
worksheet.write_formula('E51','=SUMIF(K15:K45,"Week-End", E15:E45)' )

# for project C
worksheet.write_formula('D52','=SUMIF(K15:K45,"Working Day", F15:F45)' )
worksheet.write_formula('E52','=SUMIF(K15:K45,"Week-End", F15:F45)' )

# for project D
worksheet.write_formula('D53','=SUMIF(K15:K45,"Working Day", G15:G45)' )
worksheet.write_formula('E53','=SUMIF(K15:K45,"Week-End", G15:G45)' )

# for project E
worksheet.write_formula('D54','=SUMIF(K15:K45,"Working Day", H15:H45)' )
worksheet.write_formula('E54','=SUMIF(K15:K45,"Week-End", H15:H45)' )

# for porject F
worksheet.write_formula('D55','=SUMIF(K15:K45,"Working Day", I15:I45)' )
worksheet.write_formula('E55','=SUMIF(K15:K45,"Week-End", I15:I45)' )

# for project G
worksheet.write_formula('D56','=SUMIF(K15:K45,"Working Day", J15:J45)' )
worksheet.write_formula('E56','=SUMIF(K15:K45,"Week-End", J15:J45)' )


# for calculating the total of the two columns
worksheet.write_formula('F50','=$D50+$E50')

# calculating the total hours for the month
worksheet.write_formula('F57', '=sum(F50,F51, F52, F53, F54, F55, F56)')


# calculating the total hours by row
worksheet.write_formula('L15','=IF($K15="Working Day", SUM($D15:$J15), 0)')
worksheet.write_formula('M15','=IF($K15="Week-End", SUM($D15:$J15), 0)')
worksheet.write_formula('N15','=SUM($L15:$M15)')







writer.close()
```
